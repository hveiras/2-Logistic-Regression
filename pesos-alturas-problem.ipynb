```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
```

```python
len_data = 500
data = pd.read_csv('data/alturas-pesos-mils-train.csv')[:len_data]
```

```python
from sklearn.preprocessing import MinMaxScaler
normalizeData = MinMaxScaler()
X_train  = normalizeData.fit_transform(data[['Peso', 'Altura']].values)
```

```python
labels = data['Genero'].values.copy()
labels[labels=='Hombre'] = 0
labels[labels=='Mujer'] = 1
```

```python
data_hombres = X_train[labels == 1]
data_mujeres = X_train[labels == 0]
```

```python
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import Adam, SGD
```

```python
model = Sequential()
model.add(Dense(1, input_dim=X_train.shape[1], activation='sigmoid'))
model.summary()
```

```python
# Adam lr=5 se ve inercia (len_data=500) y aprox en 100 epochs converge
# SGD lr=5 se ve que no tiene inercia y en 100 epochs queda cerca de soluci√≥n de Naive Bayes
model.compile(loss='binary_crossentropy', optimizer=Adam(lr=5), metrics=['accuracy'])
```

```python
batch_size = int(len(X_train)/1)
```

```python
from helper import PlotBoundary
```

```python
plot_boundary = PlotBoundary(X_train, labels, N=100, plots_every_batches=int(len(X_train)/batch_size))
history = model.fit(X_train, labels, epochs=100, batch_size=batch_size, 
                      verbose=0, callbacks=[plot_boundary])
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NB_clf = GaussianNB()
NB_clf.fit(X_train, data['Genero'])
```

```python
from sklearn.linear_model import LogisticRegression
log_Reg = LogisticRegression(C=1000)
log_Reg.fit(X_train, data['Genero'])
```

```python
print(NB_clf.score(X_train, data['Genero']),
      log_Reg.score(X_train, data['Genero']), 
      model.evaluate(X_train, labels, verbose=0)[1])
```
